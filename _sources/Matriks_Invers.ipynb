# Invers Matriks
Invers sendiri dapat diartikan sebagai lawan dari sesuatu (kebalikan). Invers matriks dilambangkan dengan A-1. Suatu matriks dapat dilakukan invers jika jumlah baris dan kolom pada matriks tersebut harus sama. Rumus matriks yaitu sebagai berikut:

Matriks ordo(2 x 2)
$$
\begin{equation*}
A=
\begin{bmatrix}
a&b\\
c&d
\end{bmatrix}
\end{equation*}
$$
$$
\begin{equation*}
A^-1=\frac{1}{ad-bc}
\begin{bmatrix}
d&-b\\
-c&a
\end{bmatrix}
\end{equation*}
$$

Matriks ordo (3 x 3)
$$
\begin{equation*}
A=
\begin{bmatrix}
p&q&r\\
s&t&u\\
v&w&x\
\end{bmatrix}
\end{equation*}
$$
$$
\begin{equation*}
A^-1=\frac{1}{ptx + quv + rst - vtr - wuq - xsq}
Adj A
\end{equation*}
$$

```python
import numpy as np

a=np.array([[2,-3],[-2,4]])
print('Matriks A =\n', a)
```

```python
import numpy as np

# Matriks identitas 3x3
I = np.identity(3)

# Contoh matriks A (misalnya)
A = np.array([[2, 0, 0],
              [0, 3, 0],
              [0, 0, 4]])

# Menghitung invers dari matriks A
A_inv = np.linalg.inv(A)

# Menghitung invers dari matriks identitas
I_inv = np.linalg.inv(I)

# Verifikasi persamaan (A⁻¹)⁻¹ = A
if np.allclose(I_inv, A):
    print("Persamaan (A⁻¹)⁻¹ = A terpenuhi.")
else:
    print("Persamaan (A⁻¹)⁻¹ = A tidak terpenuhi.")
```
